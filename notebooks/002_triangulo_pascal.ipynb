## Tarea 1 — Fecha de entrega: próxima clase

En esta tarea van a explorar la relación entre el triángulo de Pascal y la distribución Gaussiana. El trabajo se entrega en esta misma libreta, agregando las celdas de código y texto que consideren necesarias.

Empiecen escribiendo una función `triangulo_pascal(N)` que construya el triángulo de Pascal desde el renglón 0 hasta el renglón $N$. Verifiquen que funciona imprimiendo los primeros renglones.

Una vez que tengan el triángulo, grafiquen varios renglones como sucesiones de puntos conectados, todos en el mismo gráfico. Pueden usar una función `graficar_renglones(triangulo, renglones)` para esto. Elijan al menos 5 renglones de distintos tamaños.

El siguiente paso es normalizar cada renglón dividiendo entre su valor máximo, de modo que el pico siempre quede en 1. Escriban una función `normalizar(renglon)` que haga esto. Repitan la gráfica anterior pero con los renglones normalizados. Ahora las formas son comparables, pero cada renglón tiene un ancho diferente en el eje horizontal.

Antes de comparar con la Gaussiana, necesitan resolver el problema del eje horizontal: cada renglón tiene un número distinto de puntos, así que no se pueden superponer directamente. Escriban una función `reescalar_x(renglon, escala)` que tome las posiciones $k = 0, 1, \ldots, n$ de un renglón y las transforme a un eje centrado en cero, con un ancho controlado por el parámetro `escala`. Repitan la gráfica de los renglones normalizados pero ahora con el eje horizontal reescalado.

Finalmente, escriban una función `graficar_pascal_vs_gauss(N, escala)` que haga lo siguiente: calcule el triángulo hasta el renglón $N$, normalice los renglones seleccionados, escale el eje horizontal de cada renglón para que todos queden centrados en cero con un ancho controlado por el parámetro `escala`, y grafique encima una curva Gaussiana normalizada a pico 1, es decir $G(x) = e^{-(x-\mu)^2 / 2\sigma^2}$.

### Entregables

La libreta debe contener: las funciones `triangulo_pascal`, `normalizar`, `reescalar_x`, `graficar_renglones` y `graficar_pascal_vs_gauss` funcionando correctamente.
