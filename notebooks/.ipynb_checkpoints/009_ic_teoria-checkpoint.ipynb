```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

plt.rcParams.update(
    {
        "figure.facecolor": "white",
        "axes.facecolor": "white",
        "axes.grid": True,
        "grid.alpha": 0.3,
        "axes.spines.top": False,
        "axes.spines.right": False,
    }
)
rng = np.random.default_rng(42)
import ipywidgets as widgets
from ipywidgets import interact, interactive_output
from IPython.display import display, Markdown
```

# Semana 5A — Intervalos de Confianza (Teoría)

**Objetivo:** Entender qué es un intervalo de confianza, cómo se construye
y — sobre todo — cómo se interpreta correctamente.

**Temas:**

1. Construcción del IC paso a paso
2. σ conocida vs desconocida (z vs t)
3. Simulación de cobertura
4. Efecto del nivel de confianza

---
## 1. Construcción del intervalo de confianza

Del TLC sabemos que la media muestral se distribuye:

$$\bar{x} \;\sim\; N\!\left(\mu,\;\frac{\sigma^2}{n}\right)$$

Estandarizando: $Z = \frac{\bar{x} - \mu}{\sigma/\sqrt{n}} \sim N(0,1)$

Si queremos un intervalo que contenga $\mu$ con probabilidad $1 - \alpha$:

$$P\!\left(-z_{\alpha/2} \;\le\; \frac{\bar{x} - \mu}{\sigma/\sqrt{n}}
\;\le\; z_{\alpha/2}\right) = 1 - \alpha$$

Despejando $\mu$:

$$\boxed{\bar{x} \;\pm\; z_{\alpha/2}\;\frac{\sigma}{\sqrt{n}}}$$

donde $z_{\alpha/2}$ es el cuantil de la normal estándar
(e.g., $z_{0.025} = 1.96$ para 95% de confianza).

### Visualización de la región de confianza

```python
alpha_slider = widgets.FloatSlider(min=0.01, max=0.20, step=0.01, value=0.05, description="α (significancia)",
)

def _update(alpha_slider):
    _alpha = alpha_slider
    _conf = 1 - _alpha
    _z = stats.norm.ppf(1 - _alpha / 2)

    fig1, ax1 = plt.subplots(figsize=(9, 4))

    _x = np.linspace(-4, 4, 500)
    _pdf = stats.norm.pdf(_x)
    ax1.plot(_x, _pdf, "k-", lw=2)

    # Región central (confianza)
    _mask_center = (_x >= -_z) & (_x <= _z)
    ax1.fill_between(_x[_mask_center], _pdf[_mask_center],
                     color="steelblue", alpha=0.4,
                     label=f"Confianza = {_conf:.0%}")

    # Colas (α/2 cada una)
    _mask_left = _x <= -_z
    _mask_right = _x >= _z
    ax1.fill_between(_x[_mask_left], _pdf[_mask_left],
                     color="crimson", alpha=0.4,
                     label=f"α/2 = {_alpha/2:.3f}")
    ax1.fill_between(_x[_mask_right], _pdf[_mask_right],
                     color="crimson", alpha=0.4)

    ax1.axvline(-_z, color="crimson", ls="--", lw=1)
    ax1.axvline(_z, color="crimson", ls="--", lw=1)
    ax1.text(_z, 0.01, f"  z = {_z:.3f}", fontsize=10, color="crimson")
    ax1.text(-_z, 0.01, f"z = −{_z:.3f}  ", fontsize=10, color="crimson",
             ha="right")

    ax1.set_xlabel("Z")
    ax1.set_ylabel("Densidad")
    ax1.set_title(f"Normal estándar — Intervalo al {_conf:.0%}")
    ax1.legend(loc="upper right")
    plt.tight_layout()
    plt.show()

out = widgets.interactive_output(_update, {'alpha_slider': alpha_slider})
display(widgets.VBox([alpha_slider, out]))
```

> Al mover el slider:
> - α pequeño → colas pequeñas → $z$ grande → **IC ancho** (más seguridad)
> - α grande → colas grandes → $z$ pequeño → **IC estrecho** (menos seguridad)

## 2. σ conocida vs σ desconocida

En la práctica **casi nunca** conocemos $\sigma$ — la estimamos con la
desviación estándar muestral $s$.

Cuando usamos $s$ en lugar de $\sigma$, el estadístico ya no es $Z$ sino:

$$t = \frac{\bar{x} - \mu}{s/\sqrt{n}} \;\sim\; t_{n-1}$$

La distribución **t de Student** tiene colas más pesadas que la normal,
especialmente cuando $n$ es pequeño (pocos grados de libertad).
Esto produce ICs más anchos — penaliza la incertidumbre adicional
de no conocer $\sigma$.

```python
fig2, ax2 = plt.subplots(figsize=(9, 4))

_x = np.linspace(-5, 5, 500)
ax2.plot(_x, stats.norm.pdf(_x), "k-", lw=2, label="Normal (z)")

_colores = ["#e74c3c", "#e67e22", "#2ecc71", "#3498db"]
for gl, color in zip([3, 5, 10, 30], _colores):
    ax2.plot(_x, stats.t.pdf(_x, gl), color=color, lw=1.5,
             ls="--", label=f"t (gl = {gl})")

ax2.set_xlabel("Valor")
ax2.set_ylabel("Densidad")
ax2.set_title("Distribución t de Student vs Normal estándar")
ax2.legend()
plt.tight_layout()
plt.show()
```

> **Observa:**
> - Con **gl = 3** (n = 4), la t tiene colas notablemente más pesadas
>   → ICs mucho más anchos para compensar la incertidumbre en $s$.
> - Con **gl = 30** (n = 31), la t es casi indistinguible de la normal
>   → $s$ estima bien a $\sigma$ y la penalización es mínima.
> - Regla práctica: para $n > 30$, usar $z$ o $t$ da resultados
>   prácticamente iguales.

```python
_mu, _sigma = 25.0, 5.0
_ns = [5, 10, 30, 100]
_num_replicas = 2000

fig2b, axes2b = plt.subplots(1, len(_ns), figsize=(14, 4), sharey=True)

for ax, _n in zip(axes2b, _ns):
    _anchos_z = []
    _anchos_t = []
    for _ in range(_num_replicas):
        _muestra = rng.normal(_mu, _sigma, _n)
        _xbar = _muestra.mean()
        _s = _muestra.std(ddof=1)

        # IC con z (sigma conocida)
        _z = stats.norm.ppf(0.975)
        _ancho_z = 2 * _z * _sigma / np.sqrt(_n)
        _anchos_z.append(_ancho_z)

        # IC con t (sigma estimada)
        _t = stats.t.ppf(0.975, _n - 1)
        _ancho_t = 2 * _t * _s / np.sqrt(_n)
        _anchos_t.append(_ancho_t)

    ax.hist(_anchos_z, bins=30, density=True, color="steelblue",
            alpha=0.6, edgecolor="white", label="IC-z (σ conocida)")
    ax.hist(_anchos_t, bins=30, density=True, color="darkorange",
            alpha=0.6, edgecolor="white", label="IC-t (σ estimada)")
    ax.set_title(f"n = {_n}")
    ax.set_xlabel("Ancho del IC")
    if ax == axes2b[0]:
        ax.set_ylabel("Densidad")
    ax.legend(fontsize=7)

plt.suptitle(
    "Ancho del IC al 95%: z (σ conocida) vs t (σ estimada)",
    fontsize=13, y=1.02,
)
plt.tight_layout()
plt.show()
```

> Con $n = 5$, el IC-t es considerablemente más ancho y más variable
> que el IC-z.  Con $n = 100$, ambos son prácticamente iguales.

## 3. Simulación de cobertura

Esta es la forma más directa de entender qué significa
"intervalo al 95% de confianza".

**Interpretación correcta:**
> Si repitiéramos el proceso de muestreo muchas veces y
> construyéramos un IC cada vez, **~95% de esos intervalos
> contendrían el verdadero $\mu$**.

**Interpretación INCORRECTA:**
> ~~"Hay 95% de probabilidad de que $\mu$ esté en este intervalo."~~
>
> ($\mu$ es fijo — no tiene probabilidad. Es el intervalo el que varía
> de muestra a muestra.)

### Parámetros de la simulación

```python
n_ic_slider = widgets.IntSlider(min=5, max=100, step=5, value=20, description="Tamaño de muestra (n)",
)
num_ic_slider = widgets.IntSlider(min=20, max=200, step=10, value=100, description="Número de intervalos")
conf_slider = widgets.FloatSlider(min=0.80, max=0.99, step=0.01, value=0.95, description="Nivel de confianza")

def _update(n_ic_slider, num_ic_slider, conf_slider):
    _mu, _sigma = 25.0, 5.0
    _n = n_ic_slider
    _k = num_ic_slider
    _conf = conf_slider
    _t_crit = stats.t.ppf((1 + _conf) / 2, _n - 1)

    _medias = np.zeros(_k)
    _lo = np.zeros(_k)
    _hi = np.zeros(_k)
    _cubre = np.zeros(_k, dtype=bool)

    for i in range(_k):
        _muestra = rng.normal(_mu, _sigma, _n)
        _xbar = _muestra.mean()
        _s = _muestra.std(ddof=1)
        _margen = _t_crit * _s / np.sqrt(_n)

        _medias[i] = _xbar
        _lo[i] = _xbar - _margen
        _hi[i] = _xbar + _margen
        _cubre[i] = (_lo[i] <= _mu) & (_hi[i] >= _mu)

    _cobertura = _cubre.mean()

    fig3, ax3 = plt.subplots(figsize=(8, max(5, _k * 0.06)))

    for i in range(_k):
        _color = "steelblue" if _cubre[i] else "crimson"
        ax3.plot([_lo[i], _hi[i]], [i, i], color=_color, lw=1.2)
        ax3.plot(_medias[i], i, "o", color=_color, markersize=3)

    ax3.axvline(_mu, color="black", ls="--", lw=1.5, label=f"μ = {_mu}")
    ax3.set_xlabel("Valor")
    ax3.set_ylabel("# de muestra")
    ax3.set_title(
        f"Cobertura empírica: {_cobertura:.1%}  "
        f"(teórica: {_conf:.0%},  n = {_n},  {_k} intervalos)"
    )
    ax3.legend(loc="upper right")
    ax3.invert_yaxis()
    plt.tight_layout()
    plt.show()

out = widgets.interactive_output(_update, {'n_ic_slider': n_ic_slider, 'num_ic_slider': num_ic_slider, 'conf_slider': conf_slider})
display(widgets.VBox([n_ic_slider, num_ic_slider, conf_slider, out]))
```

> **Experimenta:**
> - Los intervalos en **rojo** son los que NO contienen $\mu$ — son
>   el error tipo I en acción.
> - Sube la confianza a 99%: menos rojos, pero intervalos más anchos.
> - Baja $n$ a 5: los intervalos se ensanchan enormemente porque
>   hay mucha incertidumbre con tan pocos datos.

## 4. Precisión vs confianza: el trade-off

Más confianza = intervalo más ancho = menos informativo.
¿Para qué quiero saber que $\mu \in [10, 40]$ con 99.99% de confianza
si eso no me dice nada útil?

El nivel 95% es una convención — no hay nada mágico en él.
La elección depende de las consecuencias de equivocarse.

```python
_mu, _sigma, _n = 25.0, 5.0, 30
_se = _sigma / np.sqrt(_n)

_niveles = np.arange(0.80, 1.00, 0.01)
_anchos = []
for _conf in _niveles:
    _z = stats.norm.ppf((1 + _conf) / 2)
    _anchos.append(2 * _z * _se)

fig4, ax4 = plt.subplots(figsize=(8, 4))
ax4.plot(_niveles * 100, _anchos, "steelblue", lw=2)
ax4.axvline(95, color="crimson", ls="--", alpha=0.7, label="95%")
ax4.set_xlabel("Nivel de confianza (%)")
ax4.set_ylabel("Ancho del IC")
ax4.set_title(
    f"Ancho del IC vs nivel de confianza  (n = {_n}, σ = {_sigma})"
)
ax4.legend()
plt.tight_layout()
plt.show()
```

> El ancho crece de forma no lineal: pasar de 95% a 99% casi
> duplica el ancho.  Es un costo creciente por cada punto
> porcentual de confianza adicional.

---
## Resumen

| Concepto | Fórmula / Idea clave |
|:---|:---|
| **IC (σ conocida)** | $\bar{x} \pm z_{\alpha/2}\;\sigma/\sqrt{n}$ |
| **IC (σ desconocida)** | $\bar{x} \pm t_{\alpha/2,\;n-1}\;s/\sqrt{n}$ |
| **Interpretación** | ~95% de los intervalos construidos contienen $\mu$ (no "95% de prob. de que $\mu$ esté aquí") |
| **Trade-off** | Más confianza → IC más ancho → menos precisión |
